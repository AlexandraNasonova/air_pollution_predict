```python
import os
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

#pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
CSV_PATH = '../../datasources/weather/'
STATION_CODE = '06344'
USE_COLUMNS = ['tavg', 'tmin', 'tmax', 'prcp', 'wdir', 'wspd', 'wpgt', 'pres']
DATE_COLUMN = 'date'
```

```python
df = pd.read_csv(CSV_PATH + STATION_CODE + '.csv', index_col='date', parse_dates=True)
```

```python
df.head()
```

```python
df.describe()
```

```python
df = df[USE_COLUMNS]
```

```python
for col in USE_COLUMNS:
    axs = plt.gca()
    df.plot(xlabel=DATE_COLUMN, ylabel=col, title=col,  y=col, c='tab:blue', figsize=(21,7), ax=axs)
    plt.show()
```
