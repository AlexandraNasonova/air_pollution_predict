1) для загрязнителя с наибольшим количеством файлов в каждом датасете создать колонку localid записав туда локальный id, извлеченный из названия файла. 
<br>2) Затем нам необходимо объединить все файлы для выбранной страны в один единый датасет. 
<br>3) И затем из полученного датасета извлечь таблицу соответствия localid и EoI-кодов для выбранной страны
<br>4) Повторить все то же самое для других стран, данные которых планируется использовать в работе

В связи с неудобством работы с датасетами очень большого размера (больше нескольких гигабайт), времязатратностью написания и исполнения кода для работы с такими датасетами, в качестве возможных альтернатив видятся два возможных решения:
<br>
1) Проверить несколько других стран и посмотреть будет ли в них такая же проблема с localid. Если получится найти страны где есть привязка localid к EoI-кодам, использовать их для дальнейшей работы.
<br>2) Пересмотреть подход к работе и выбрать страну (или страны) с небольшим количеством данных для каждого загрязнителя, для того чтобы их можно было объединить в датасеты не слишком большого размера.
